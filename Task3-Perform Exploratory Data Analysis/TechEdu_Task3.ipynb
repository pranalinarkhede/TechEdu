## Cluster of Annual Income (k) V/s Spending Score

```python
plt.figure(figsize=(12,5))
plt.scatter("Annual Income (k$)","Spending Score (1-100)",data=Mall_data,c="Spending Score (1-100)",s=75)
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title('Clustering of Annual Income(k) Vs Spending Score(1-100)')
plt.legend()
plt.grid()
plt.show()
```

From the above graph it can be observed that there are 5 clusters.
But lets verify it by Elbow method. 

```python
#Training the Model by taking different clusters ranging between 1 to 10 that will help us to find the Elbow point
from sklearn.cluster import KMeans
kmeans=[]
for i in range(1,20):
    KMNS=KMeans(n_clusters=i,random_state=0)
    KMNS.fit(Mall_data.iloc[:,3:4])
    kmeans.append(KMNS.inertia_)
```

```python
#find the Elbow point by plotting the trained model
plt.plot(range(1,20),kmeans,color="blue",marker="o")
plt.xlabel("No. of Clusters")
plt.ylabel("Within Cluster Sum of Square")
plt.title("The Elbow Method")
plt.grid(linestyle='-.')
plt.show()
```

The curve shows similar behaviour after x=5.0, hence it is clear that no. of clusters are 5.
